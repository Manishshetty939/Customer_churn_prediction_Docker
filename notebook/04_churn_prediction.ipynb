### Loading libraries and defining churn

```python
import pandas as pd 
import numpy as np 
from datetime import timedelta 

rfm = pd.read_csv(r"C:\Users\Deepika Hegde\customer_sales_forecasting\data\rfm_processed.csv")
snapshot_date = pd.to_datetime('2011-12-10')

# adding churn label 
rfm['Churn'] =  np.where(rfm['Recency']>180,1,0) 
rfm['Churn'].value_counts()
```

### Feature Selection

```python
X = rfm[['Recency', 'Frequency', 'Monetary']]
y = rfm['Churn']
```

### Train-Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42) 
```

### Train Classifier (Random Forest)

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,confusion_matrix 

model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train,y_train)

y_pred = model.predict(X_test)
print(classification_report(y_test,y_pred))
```

### Visualize Results

```python
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.metrics import ConfusionMatrixDisplay 

ConfusionMatrixDisplay.from_estimator(model,X_test,y_test) 
plt.title('Random Forest Classifier Confusion Matrix')
plt.show()
```

###  Feature Importance

```python
importances = model.feature_importances_ 
feat_names = X.columns 

sns.barplot(x=importances,y=feat_names) 
plt.title('Feature Importances for Churn Prediction')
plt.show()
```
