### Revenue by segment

```python
import pandas as pd 
import matplotlib.pyplot as plt

rfm = pd.read_csv(r"C:\Users\Deepika Hegde\customer_sales_forecasting\data\rfm_churn_pred.csv")
# Example: Revenue by Segment
segment_rev = rfm.groupby('Segment')['Monetary'].sum().sort_values(ascending=False)
segment_rev.plot(kind='bar', title='Revenue by Segment', figsize=(8,5))
plt.ylabel("Total Revenue")
plt.tight_layout()
plt.savefig(r'C:\Users\Deepika Hegde\customer_sales_forecasting\report')
```
